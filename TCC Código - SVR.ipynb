```python
import pandas as pd

ds = pd.read_csv("./Base de dados/GOB_radiation_2014-07.tab", sep='\t', engine='python')
# print(ds.title)
# print(ds.data.head())
ds
```

```python
ds = ds[ds.index < 2*24*60]
```

Cleaning the code (getting only 6AM to 6PM)

```python
ds.time = pd.to_datetime(ds['Date/Time'])
ds.time
```

```python
clean_ds = ds[ds.time.dt.strftime('%H:%M').between('06:00','18:00')]
clean_ds
```

get into the statistics fun

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn import datasets, metrics
from mealpy.bio_based import SMA #trocar por hho
import numpy as np
```

```python
x = clean_ds[['T2 [°C]', 'RH [%]', 'PoPoPoPo [hPa]']]
x.fillna(0)
x
```

```python
y = clean_ds[['SWD [W/m**2]']]
y.fillna(0)
y
```

```python
y = y.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)
y
```

separate training and test data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
x_train, x_test, y_train, y_test
```

```python
sc = StandardScaler()
sc.fit(x_train)
X_train_std = sc.transform(x_train)
X_test_std = sc.transform(x_test)
```

```python
KERNEL_ENCODER = LabelEncoder()
KERNEL_ENCODER.fit(['linear', 'poly', 'rbf', 'sigmoid'])
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

svr = RandomForestRegressor(n_estimators=30)
# Fit the model
svr.fit(X_train_std, y_train.values.ravel())
# Make the predictions
y_predict = svr.predict(X_test_std)
# Measure the performance
x_print = metrics.r2_score(y_test.values.ravel(), y_predict)
print(x_print)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_validate
def fitness_function(solution, flg = None):
#     if kernel belongs to 0 - 0.99 ==> 0       ==> linear
#                           2 - 2.99 ==> 2
#                           3 - 3.99 ==> 3      ==> sigmoid

#     kernel_encoded = round(solution[0])
#     c = solution[1]
#     kernel_decoded = KERNEL_ENCODER.inverse_transform([kernel_encoded])[0]
    kernel_list = ['linear', 'poly', 'rbf', 'sigmoid']
#     svc = SVC(C=c, random_state=1, kernel=kernel_decoded, max_iter=100)
    reg = SVR(C = round(solution[1]), kernel=kernel_list[round(solution[0])])
#     print(kernel_list[solution[0]%5])
    # Fit the model
    accuracy = cross_validate(reg, X_train_std, y_train.values.ravel(), cv=5, scoring='max_error')
    result = np.array(accuracy['test_score']).mean()
    
    if(flg == 'run'):
        print(accuracy)
        reg.fit(X_train_std, y_train.values.ravel())
        return reg
    
    return abs(result)
    # Make the predictions
#     y_predict = svc.predict(X_test_std)
    # Measure the performance
#     x_print = cross_val_score(reg, X_train_std, y_train.values.ravel(), cv=3).mean()
#     print(solution, x_print)
#     return x_print

problem = {
    "fit_func": fitness_function,
    "lb": [  0,  1e0],
    "ub": [  3, 1e+4],
    "minmax": "min"
}
```
